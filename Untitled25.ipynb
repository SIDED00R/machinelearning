```python
from keras_facenet import FaceNet
embedder = FaceNet()
```

```python
import matplotlib.image as mpimg 
import cv2
import numpy as np
import pandas as pd
```

```python
def load_train_data(n):
  lst = []
  for i in range(1,n+1):
    filename = str(i)+'.jpg'
    x = mpimg.imread('/content/drive/MyDrive/Colab Notebooks/facenet/train/'+filename)
    x = cv2.resize(x,(160,160))
    lst.append(x)
  df = np.array(lst)
  y_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/facenet/train/y_train.csv',header=None)
  y_train = y_train.values
  return df,y_train
```

```python
def load_test_data(n):
  lst = []
  for i in range(1,n+1):
    filename = str(i)+'.jpg'
    x = mpimg.imread('/content/drive/MyDrive/Colab Notebooks/facenet/Test/'+filename)
    x = cv2.resize(x,(160,160))
    lst.append(x)
  df = np.array(lst)
  y_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/facenet/Test/y_test.csv',header=None)
  y_test = y_test.values
  return df,y_test
```

```python
X_train_orig,y_train = load_train_data(24)
X_test_orig,y_test = load_test_data(7)
print('X_Train data shape=',X_train_orig.shape)
print('X_Test data shape=',X_test_orig.shape)
print('y_Train data shape=',y_train.shape)
print('y_Test data shape=',y_test.shape)
```

```python
X_train = embedder.embeddings(X_train_orig)
X_test = embedder.embeddings(X_test_orig)
print('Train embed shape=',X_train.shape)
print('Test embed shape=',X_test.shape)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
#Predicted faces
model.predict(X_test)
```

```python
#True faces
y_test
```

```python
model.score(X_test,y_test)
```

```python
def predict(path, filename):
  x = mpimg.imread(path+'/'+filename)
  x = cv2.resize(x,(160,160))
  x = x.reshape(1,160,160,3)
  x = embedder.embeddings(x)
  if(max(model.predict_proba(x)[0])<0.25):         #0.25 is the threshold
    print("Face not found !")
    return
  else:
    return model.predict(x)
```
