```python
%cd /
from glob import glob

img_list = glob('/content/yolov5/data/images/train/*.jpg')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/yolov5/data/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/yolov5/data/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
#yaml ÌååÏùº ÏàòÏ†ï
import yaml

with open('/content/yolov5/data/coco.yaml','r') as f:
  data = yaml.load(f,Loader=yaml.FullLoader)

print(data)

data['train'] = '/content/yolov5/data/train.txt'
data['val'] = '/content/yolov5/data/val.txt'
data['nc'] = 4
data['names'] = ['lefteye','righteye','nose','mouth']

with open('/content/yolov5/data/coco.yaml', 'w') as f:
  yaml.dump(data, f)

  print(data)
```
