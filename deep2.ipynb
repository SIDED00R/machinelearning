```python
import torch
import numpy as np
import matplotlib.pyplot as plt

x_data=[1,2,3]
y_data=[2,4,6]
plt.plot(x_data,y_data)
plt.xlabel('hours')
plt.ylabel('points')
plt.show()
```

```python
def forward(x):
  return x*w
def loss(x,y):
  y_pred=forward(x)
  return(y_pred-y)*(y_pred-y)
w_list=[]
mse_list=[]
for w in np.arange(0,4.1,0.1):
  l_sum=0
  for x_val,y_val in zip(x_data,y_data):
    y_pred_val=forward(x_val)
    l_sum=loss(x_val,y_val)
  w_list.append(w)
  mse_list.append(l_sum/3)
plt.plot(w_list,mse_list)
plt.ylabel('loss')
plt.xlabel('w')
plt.show()
```

```python
w=1
def gradient(x,y):
  return 2*x*(x*w-y)
for epoch in range(100):
  l=0
  for x_val,y_val in zip(x_data,y_data):
    grad=gradient(x_val,y_val)
    w=w-0.01*grad
    l=l+loss(x_val,y_val)
  mse=l/3
  print("progress: ",epoch, "w = ",mse)
print("test ", 4,forward(4))
```
